### 2. De falso color a la imagen en blanco y negro “paisaje.jpg” componiendo las capas RGB
* El cielo debe quedar azul y la arena debe ser amarilla

````python
img_gray = cv2.imread('../Imagenes/paisaje.jpg', 0)/255

img_out = np.zeros((img_gray.shape[0], img_gray.shape[1], 3))

img_out[..., 0] = img_gray
img_out[..., 1] = img_gray
img_out[..., 2] = img_gray

img_out[0:68, :, 0] = img_out[0:68, :, 0] - 10/255
img_out[0:68, :, 1] = img_out[0:68, :, 1] + 50/255
img_out[0:68, :, 2] = img_out[0:68, :, 2] + 70/255

img_out[68:, :, 0] = img_out[68:, :, 0] + 50/255
img_out[68:, :, 1] = img_out[68:, :, 1] + 50/255
img_out[68:, :, 2] = img_out[68:, :, 2]


plt.figure(figsize=(15,8))
plt.imshow(img_out, vmin=0, vmax=1)
plt.show()
````

### 6. Cambie los matices de los autos opacos en el espacio HSV, luego muestre una comparacion entre la entrada y la salida
* Trabaje con al menos con las capas H y S del modelo HSV

````python
autos=cv2.imread("../Imagenes/autos.jpg")
plt.imshow(autos)
plt.show()
````

````python
autosHSV=cv2.cvtColor(autos, cv2.COLOR_RGB2HSV)

hue = autosHSV[:,:,0].copy()
sat = autosHSV[:,:,1].copy()
val = autosHSV[:,:,2].copy()

mask = sat > 100

plt.figure(figsize=(16,8))
plt.subplot(2,3,1)
plt.imshow(hue)
plt.axhline(y=25, color='r', linestyle='dashed', linewidth=2)
plt.subplot(2,3,2)
plt.imshow(sat)
# plt.axhline(y=25, color='r', linestyle='dashed', linewidth=2)
plt.axhline(y=150, color='r', linestyle='dashed', linewidth=2)
plt.subplot(2,3,3)
plt.imshow(val)
plt.subplot(2,3,4)
plt.plot(hue[25,:])
plt.subplot(2,3,5)
plt.plot(sat[150,:])
plt.subplot(2,3,6)
plt.imshow(mask)
plt.show()
````

````python
aux = mask * hue
aux2 = mask * sat

autosHSV[:,:,0] = aux
autosHSV[:,:,1] = aux2
autosHSV[:,:,2] = val*mask

autosT_color=cv2.cvtColor(autosHSV, cv2.COLOR_HSV2RGB)

plt.figure(figsize=(16,8))
plt.subplot(141)
plt.imshow(autosHSV[:,:,0])
plt.subplot(142)
plt.imshow(autosHSV[:,:,1])
plt.subplot(143)
plt.imshow(autosHSV[:,:,2])
plt.subplot(144)
plt.imshow(autosT_color)
plt.show()
````

````python
all_a = val<252
dark_a = np.bitwise_xor(all_a,mask)
plt.imshow(dark_a)
plt.show()
````

````python
hue_d = hue * dark_a
sat_d = sat * dark_a
index = dark_a>0

hue_d[index] += 40
sat_d[index] += 100

autosHSV[:,:,0] = hue_d
autosHSV[:,:,1] = sat_d
autosHSV[:,:,2] = val

autosT_oscuro=cv2.cvtColor(autosHSV, cv2.COLOR_HSV2RGB)

plt.figure(figsize=(16,8))
plt.subplot(141)
plt.imshow(autosHSV[:,:,0])
plt.subplot(142)
plt.imshow(autosHSV[:,:,1])
plt.subplot(143)
plt.imshow(autosHSV[:,:,2])
plt.subplot(144)
plt.imshow(autosT_oscuro)
plt.show()
````

````python
autosHSV[:,:,0] = hue_d + aux
autosHSV[:,:,1] = sat_d + aux2
autosHSV[:,:,2] = val

autosT_final = cv2.cvtColor(autosHSV, cv2.COLOR_HSV2RGB)

plt.figure(figsize=(16,8))
plt.subplot(141)
plt.imshow(autosHSV[:,:,0])
plt.subplot(142)
plt.imshow(autosHSV[:,:,1])
plt.subplot(143)
plt.imshow(autosHSV[:,:,2])
plt.subplot(144)
plt.imshow(autosT_final)
plt.show()
````

### 8. Use OpenCV para activar y visualizar la cámara
* Muestre la visualización de la camará

````python
cap = cv2.VideoCapture(0)

while True:
    _, frame= cap.read()

    frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    cv2.imshow('primera app en tiempo real', frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
````

* Segmente un color en la visualizacion de la camara

````python
cap = cv2.VideoCapture(0)

while True:
    _, frame= cap.read()

    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)

    lower = np.array([0, 90, 50])
    upper = np.array([60, 220, 180])

    mask = cv2.inRange(hsv, lower, upper)

    res = cv2.bitwise_and(frame, frame, mask = mask)

    cv2.imshow('frame', frame)
    cv2.imshow('mask', mask)
    cv2.imshow('result', res)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
````
